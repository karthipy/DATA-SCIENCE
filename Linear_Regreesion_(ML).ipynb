```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

```

```python
data = pd.read_csv("BIKE DETAILS.csv")
data
```

```python
print(data["name"].unique())
```

```python
yamaha = data[data["name"].str.contains("Yamaha FZ S", case=False, na=False)]
print(yamaha)
```

```python
bike = yamaha[["year","selling_price"]]
print(bike)
```

```python
plt.plot(bike["year"],bike["selling_price"])
plt.title("Bike Selling Price")
plt.xlabel("Year")
plt.ylabel("Selling Price")
plt.show()
```

```python
x = bike[["year"]]
y = bike[["selling_price"]]
model = LinearRegression()
model.fit(x,y)
```

```python
price = model.predict([[2028]])
print("YAMAHA FZ S PREDICTED PRICE IN 2028:",price[0])
```
