```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("TN.csv")

top10 = data.sort_values(
    'Population\nCensus\n2011-03-01',
    ascending=False
).head(10)

plt.figure(figsize=(10,6))
plt.plot(
    top10['District'],
    top10['Population\nCensus\n2011-03-01']
)

plt.title("Tamil Nadu Population")
plt.xlabel("Population")
plt.ylabel("District")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("deliveries.csv")
data=pd.read_csv("matches.csv")


top10_winners = data['winner'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(top10_winners.index, top10_winners.values)
plt.title("IPL WINNERS")
plt.xlabel("WINNER")
plt.ylabel("Number of Wins")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("BIKE.csv")

bike_Brand = data["Brand"].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.pie(
    bike_Brand.values,
    labels = bike_Brand.index,
    autopct="%1.1f%%"
)

plt.title("Top Sales Bike Brands")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("laptop.csv")

plt.figure(figsize=(18,6))
plt.hist(data["Price"],bins=10,edgecolor="black")

plt.title("LAPTOP PRICE DISTRIBUTION")
plt.xlabel("PRICE")
plt.ylabel("NUMBER OF LAPTOPS")
plt.xticks(data["Price"][::200], rotation=45)

plt.tight_layout()
plt.show()
```
